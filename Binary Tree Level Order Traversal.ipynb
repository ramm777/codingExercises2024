Given the root of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).

Input: root = [3,9,20,null,null,15,7]
Output: [[3],[9,20],[15,7]]

Input: root = [1]
Output: [[1]]

Input: root = []
Output: []



```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
            self.val = val
            self.left = left
            self.right = right
```

```python
def levelOrder(root):

    if not root:
        return []
    
    queue  = [root]
    result = []
        
    while queue:
        level_size = len(queue)
        subtrees_list = []  # a list of subtrees
        
        # Iterate over all items at level, level_size can be larger that 2. Append all childs in the level to the queue. 
        for _ in range(level_size):
            node = queue.pop(0)
            subtrees_list.append(node.val)
            
            if node.left:  queue.append(node.left)
            if node.right: queue.append(node.right)
        
        # Append a list of subtrees as a separate list to the results list
        result.append(subtrees_list)
    
    return result
```

```python
# [3,9,20,None,None,15,7]
root = TreeNode(3, TreeNode(9), TreeNode(20))
root.right.left  = TreeNode(15)
root.right.right = TreeNode(7)

root.left.left  = TreeNode(None)
root.left.right = TreeNode(3)

# Your provide the whole tree for the function
levelOrder(root)
```
