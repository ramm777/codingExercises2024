Given an integer array nums where the elements are sorted in ascending order, convert it to a height-balanced binary search tree.

Conditions for Height-Balanced Binary Tree:

- The difference between the heights of the left and the right subtree for any node is not more than one.
- The left and the right subtrees are balanced.



- AK: Looks like Leetcode uses Level Order Traverse (i.e. travel) for the displaying trees numerically

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
            self.val = val
            self.left = left
            self.right = right


def sortedArrayToBST(nums):
    if not nums:
        return None
    
    # You need an inner function to call it again (recursevely) for left/right subnodes of a TreeNode class. 
    def createNode(left_id, right_id):
        
        if left_id > right_id:
            return None
        
        # (2) Find the mid element and place it as the root. The "//" is floor division for indices. 7//2=3, but -7//2=-4 (rounds towards negative infinity)
        mid_id = (left_id + right_id) // 2                
        node = TreeNode(nums[mid_id])               
        
        # (3) Create Left/Right of a TreeNode by specifying updated indices from both sides of the root in a list
        node.left  = createNode(left_id,    mid_id - 1) # right_id is decreasing towards left_id up to "left_id > right_id"
        node.right = createNode(mid_id + 1, right_id)   # left_id is increasing towards right_id up to "left_id > right_id"
        return node
    
           # (1) Take the first and last indices of a sorted list. The mid must be a root as sorted to height balanced 
    return createNode(0, len(nums) - 1)	
```

```python
nums = [-10,-3,0,5,9]
print(sortedArrayToBST(nums))
```
