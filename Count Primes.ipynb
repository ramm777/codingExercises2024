 Given an integer n, return the number of prime numbers that are strictly less than n.
 
 A prime number is a number that can only be divided by itself and 1 without remainders

 



https://www.youtube.com/watch?v=klcIklsWzrY

```python
def count_primes(n: int) -> int:
    if n <= 2:
        return 0
    
    # Initialize a list, all as primes, then eliminate nonprimes and multiples as you go along is_prime list
    is_prime = [True] * n

    # Base cases, known
    is_prime[0] = is_prime[1] = False  
    
    # Sieve of Eratosthenes. Start from 2 as we know that it is the smallest prime number. 
    for i in range(2, n):
        if is_prime[i]:
            # Eliminate current prime number's multiples, as they become non-prime (eg 4/2, 6/2, 8/2,...), by specifying step "i" in range()
            for j in range(i * i, n, i):
                is_prime[j] = False
    
    # Count the number of prime numbers
    return sum(is_prime)

n = 100
print(f"Number of primes less than {n}: {count_primes(n)}")
```
