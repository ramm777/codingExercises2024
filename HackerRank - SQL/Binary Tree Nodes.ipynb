```python
SELECT
    N, 
    CASE
        WHEN P IS NULL THEN 'Root' 
        WHEN N IN 
            (SELECT DISTINCT P 
            FROM BST
            WHERE P IS NOT NULL) THEN 'Inner' -- By definition, if the node is Parent, it is an Inner node
         ELSE 'Leaf'
      END AS Node_Type
FROM BST
ORDER BY N;   
```
