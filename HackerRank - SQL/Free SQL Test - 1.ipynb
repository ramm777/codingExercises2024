You are a data analyst working for a large US based supoerstore. You have been granted access to historic sales database.

- Table orders contain 'orderId', 'customerId', 'productId', 'orderDate', 'value'
- Table returns contains 'orderId', 'returned' which is integer. 

Your manager would like to see the 'orderId', 'customerId', 'productId' for the last 15 returned orders, based on the 'orderDate' column. If two returned orders were ordered on the same day, sort by value, showing the highest order first. Note: an order can span multiple rows in the 'orders' table, but each row in an order will have the same 'orderId'. 

```python
SELECT DISTINCT o.orderId, o.customerId, o.productId
FROM orders o INNER JOIN returns r 
ON o.orderId = r.orderId  
WHERE r.returned = 1
ORDER BY o.orderDate DESC, o.value DESC
LIMIT 15;
```
