-- Which store had the highest number of separate purchases during April 2021? 
SELECT store_id
FROM (
    SELECT store_id, COUNT(purchase_id) AS purchase_count
    FROM purchase
    WHERE purchase_date >= '2021-04-01' AND purchase_date < '2021-05-01'
    GROUP BY store_id
    ORDER BY purchase_count DESC
    LIMIT 1 );
