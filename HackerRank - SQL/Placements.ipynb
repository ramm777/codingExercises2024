```python
SELECT grouped.Name
FROM 
    (SELECT s.ID, s.Name, p.Salary as Salary, f.Friend_ID, p1.Salary AS Friend_Salary
    FROM Students s
    INNER JOIN Packages p
    ON s.ID = p.ID
    INNER JOIN Friends f
    ON s.ID = f.ID
    INNER JOIN Packages p1
    ON 
    p1.ID = f.Friend_ID
    ORDER BY Friend_Salary ASC) grouped
WHERE grouped.Friend_Salary > grouped.Salary;   
```
