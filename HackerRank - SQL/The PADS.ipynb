- Query an alphabetically ordered list of all names in OCCUPATIONS, immediately followed by the first letter of each profession as a parenthetical (i.e.: enclosed in parentheses). 

- Query the number of ocurrences of each occupation in OCCUPATIONS. Sort the occurrences in ascending order, and output them in the following format. The results of the second query are ascendingly ordered first by number of names corresponding to each profession, and then alphabetically by profession.





```python
SELECT CONCAT(Name, '(', LEFT(Occupation, 1), ')') 
FROM OCCUPATIONS
ORDER BY name; 
--
SELECT 
    CONCAT('There are a total of ', COUNT(Occupation), ' ', 
    LOWER(Occupation), 
    CASE 
        WHEN COUNT(Occupation) > 1 THEN 's.' 
        ELSE '.' 
    END) AS Message
FROM 
    OCCUPATIONS
GROUP BY 
    Occupation
ORDER BY 
    COUNT(Occupation) ASC, Occupation;
```
