SELECT CAST( ROUND(LAT_N, 4) AS DECIMAL(10,4) ) AS lat_n_median
FROM 
    (SELECT 
         ROW_NUMBER() OVER (ORDER BY LAT_N) AS row_num,
         (SELECT COUNT(*) FROM STATION) AS total_count,
          LAT_N    
     FROM STATION) grouped
WHERE grouped.row_num IN (  (grouped.total_count + 1) / 2, (grouped.total_count + 2) / 2 );
