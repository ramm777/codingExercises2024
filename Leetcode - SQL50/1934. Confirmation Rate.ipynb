SELECT s.user_id, 
        CASE
           WHEN c.confirmed_count / c.total_count IS NULL THEN 0
           ELSE c.confirmed_count / c.total_count
        END AS confirmation_rate   
FROM Signups s LEFT JOIN( SELECT user_id,
                                 COUNT(user_id) AS total_count,
                                 SUM(CASE WHEN action = 'confirmed' THEN 1 ELSE 0 END) AS confirmed_count
                          FROM Confirmations
                          GROUP BY user_id) c 
ON s.user_id = c.user_id;
