```python
# This solution is simple but long. Importantly, nums1 has length of m+n, so there is n extra space already in nums1 to use for the nums2
# You iterate over nums2, and force set (replace) each value from nums2 to nums1 starting from m index
# sort() has O(k*log(k)) time complexity, where k is the length of total array m+n

nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3

def merge(nums1, m, nums2, n):
    for j in range(n):
        nums1[m+j] = nums2[j]
    nums1.sort()

    return nums1

print(merge(nums1, m, nums2, n))    
```

```python
# This solution is O(m+n), faster but more difficult to remember and implement. 
# We iterate through both m and n lists stepwise, compare items to each other from both lists, and place the biggest one to the end on m+n list. 
# It's important that you move (backwards) only through the list that has higher item, only move that one. k is moved irrespectively.  
# At some point, k will traverse into m, and change correctly. 

nums1 = [1,2,3,0,0,0] # m+n length
m = 3
nums2 = [2,5,6]
n = 3

def merge(nums1, m, nums2, n):
    i = m - 1
    j = n - 1
    k = m + n - 1
    
    while j >= 0:
        if i >= 0 and nums1[i] > nums2[j]:
            nums1[k] = nums1[i]
            i -= 1
        else:
            nums1[k] = nums2[j]
            j -= 1
        k -= 1

```
