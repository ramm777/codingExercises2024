```python
nums = [1,0,1,0,3,12]

# There are only two items: zeros and non-zeros
# fast moves through all items in a list
# slow only moves conditionally when fast is not equal to 0

slow = 0
for fast in range(len(nums)): 
    if nums[fast] != 0: 
        # Move non-zeros to the earlier (front) => nums[fast] = nums[slow]
        # Move zeros to the later (end)                 => nums[slow] = nums[fast]
        nums[slow], nums[fast] = nums[fast], nums[slow]

        # Because you did swapping, slow ind must move forward
        slow += 1

nums
```

```python
nums = [1,0,1,0,3,12]

# You need to find nonzeros first to send them to the earlier (swap)
zeros = 0
for nonzeros in range(len(nums)):
    if nums[nonzeros] != 0: 
        nums[nonzeros], nums[zeros] = nums[zeros], nums[nonzeros]
        zeros += 1

nums
```
