```python
from typing import List

def rotate(nums: List[int], k: int) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """

    # If rotations k more than the length of nums, it means we rotate "back to the original" + "remainder" number of times. Thus, consider only remainder rotation. 
    if k > len(nums): 
        k = k % len(nums) 

    if  len(nums) == 1 or k == 0:
        pass
    else:     
        nums[:] = nums[-k:] + nums[:len(nums) - k]
    
rotate(nums, k)
nums
```
