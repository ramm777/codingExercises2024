- It is solved using random.randint() and Fisher-Yates algorithm 
- The only way is to learn this by heart. Especially the fact that it makes permuations equally likely. 

```python
# Runtime is not great here because of deepcopy. You can make a shallow copy instead [:], as integers are immutable anyway. 
# However, if not integers, or mutable items in an array, then better to use a deepcopy.  

from typing import List
import copy
import random

class Solution:

    def __init__(self, nums: List[int]):
        self.nums = copy.deepcopy(nums)       # = nums[:]
        self.nums_orig = copy.deepcopy(nums)  # = nums[:]

    def reset(self) -> List[int]:
        self.nums = copy.deepcopy(self.nums_orig) # = self.nums_orig[:]
        return self.nums_orig

    def shuffle(self) -> List[int]:
        
        for i in range(len(self.nums)):
            # Pick a random index from i to the end
            swap_idx = random.randint(i, len(self.nums) - 1)  
            
            # Swap elements at i and swap_idx
            self.nums[i], self.nums[swap_idx] = self.nums[swap_idx], self.nums[i]
        
        return self.nums


nums = [1, 2, 3]
obj = Solution(nums)
param_1 = obj.reset()
param_2 = obj.shuffle()
```
