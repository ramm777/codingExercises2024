```python
from typing import List

#nums = [2,2,3,3,1,5,5]
nums = [1]
#nums = [4,1,2,1,2]
```

```python
duplicate_counts = {}
for num in nums: 
    if num in duplicate_counts.keys(): 
        duplicate_counts[num] = duplicate_counts[num] + 1
    else: 
        duplicate_counts[num] = 1

duplicate_counts
```

```python
for key, value in duplicate_counts.items():
    if value < 2:
        print(key) 
```

```python
def singleNumber(nums: List[int]) -> int:

    duplicate_counts = {}
    for num in nums: 
        if num in duplicate_counts.keys(): 
            duplicate_counts[num] = duplicate_counts[num] + 1
        else: 
            duplicate_counts[num] = 1

    for key, value in duplicate_counts.items():
        if value < 2:
            return key 
        

singleNumber(nums)        
```
