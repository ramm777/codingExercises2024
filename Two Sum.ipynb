```python
nums = [7,3,2,4] 
target = 6
nums
```

```python
# Brute force O(n^2) - two loops
for i in range(len(nums)):
    for j in range(i+1, len(nums)): 
        if nums[i] + nums[j] == target:
            print([i,j])
```

```python
# O(nlog(n)) - one loop, seen(num:index)

def TwoSum(nums, target): 
    seen = {}
    for i in range(len(nums)):
        
        diff = target - nums[i]
        if diff in seen.keys(): 
            return [seen[diff], i]
        else: 
            seen[nums[i]] = i
            
TwoSum(nums, target)   
```

```python
# If you record diff and seen before comprison, then the current num will be compared with itself
# This is also a question in interview. Think more about this situation.  

seen = {}
for i in range(len(nums)):
    
    diff = target - nums[i]
    seen[nums[i]] = i
    
    if diff in seen.keys(): 
        print([seen[diff], i])
        break

seen    
```
